## Memeriksa Hipotesis

**Apakah terdapat korelasi antara memiliki anak dengan probabilitas melakukan gagal bayar pinjaman?**

```python
pivot_table_children = new_df.pivot_table(index='children', columns='debt' ,values='total_income', aggfunc='count')
pivot_table_children[1] = pivot_table_children[1].fillna(0)
pivot_table_children['percentage_debt'] = pivot_table_children[1]/(pivot_table_children[1]+pivot_table_children[0])
pivot_table_children['percentage_debt'] = pivot_table_children['percentage_debt'].apply("{:,.2%}".format)
pivot_table_children
```

Untuk nasabah yang tidak memiliki anak memiliki presentasi tidak bayar sangat rendah dibandingkan yang lain. Hal ini mungkin bisa terjadi karena masih belum banyak tanggungan seperti tanggungan anak. Sehingga bisa melunasi pinjaman dengan lancar. Namun, hal yang tidak terduga adalah nasabah dengan anak 5 yang bisa melunasi hutang. Apakah ini disebabkan karena pendapatan mereka yang besar?

**Apakah terdapat korelasi antara status keluarga dengan probabilitas melakukan gagal bayar pinjaman?**

```python
pivot_table_family= new_df.pivot_table(index='family_status', columns='debt' ,values='total_income', aggfunc='count')
pivot_table_family['percentage_debt'] = pivot_table_family[1]/(pivot_table_family[1]+pivot_table_family[0])
pivot_table_family['percentage_debt'] = pivot_table_family['percentage_debt'].apply("{:,.2%}".format)
pivot_table_family
```

Untuk nasabah yang sudah janda atau duda, memiliki persentase lebih rendah untuk tidak bayar. Hal ini mungkin bisa terjadi karena sebagian besar dari mereka itu sudah hidup sendiri dan tidak memiliki tanggungan untuk keluarga atau anak. Yang paling tertinggi adalah nasabah dengan status unmarried atau belum menikah.

## Kesimpulan Umum

Analisis dimulai dengan kita membuat hipotesis. Hipotesis ini meliputi:
1. Seseorang gagal dalam membayar pinjaman karena ada pengeluaran lain untuk istri dan anaknya
2. Seseorang gagal dalam membayar pinjaman karena pendapatannya yang ternyata tidak cukup.
3. Perbedaan tujuan dalam melakukan pinjaman bisa menjadi faktor seseorang gagal dalam melakukan pinjaman.

Sebelum menguji hipotesis tersebut, kita terlebih dahulu melakukan pra-pemrosesan pada data, diantaranya:
1. Mengubah nilai pada kolom education menjadi huruf kecil semua.

3. Merubah nilai negatif pada kolom days_employed menjadi nilai positif.
4. Menghapus nilai 0 pada komo dob_years.
5. Menghapus nilai unik XNA pada kolom gender.
6. Menghapus nilai duplikat.

Sehingga, membuat kita bisa mudah dalam melunasi peminjaman.
